# Time Series Analysis for IOTA Dataset (without considering the Entities)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import ast

# Connect to the SQLite database
conn = sqlite3.connect('iotaDB.db')

# Query the transaction data
query_transactions = "SELECT timestamp, input_addresses_x, output_addresses_y, input_amounts_x, output_amounts_y FROM Transactions"
transactions_df = pd.read_sql_query(query_transactions, conn)

# Close the database connection
conn.close()

# Convert timestamp to datetime
transactions_df['timestamp'] = pd.to_datetime(transactions_df['timestamp'])

# Function to calculate balances for each address
def calculate_balances(transactions):
    address_balances = {}

    for _, tx in transactions.iterrows():
        input_addresses = ast.literal_eval(tx['input_addresses_x'])
        output_addresses = ast.literal_eval(tx['output_addresses_y'])
        input_amounts = ast.literal_eval(tx['input_amounts_x'])
        output_amounts = ast.literal_eval(tx['output_amounts_y'])

        for addr, amount in zip(input_addresses, input_amounts):
            if addr in address_balances:
                address_balances[addr] -= amount
            else:
                address_balances[addr] = -amount

        for addr, amount in zip(output_addresses, output_amounts):
            if addr in address_balances:
                address_balances[addr] += amount
            else:
                address_balances[addr] = amount

    return address_balances

# Calculate balances
address_balances = calculate_balances(transactions_df)

# Convert the address balances to a DataFrame
balances_df = pd.DataFrame(list(address_balances.items()), columns=['address', 'balance'])
balances_df = balances_df[balances_df['balance'] > 0]  # Filter out addresses with zero or negative balance

# Add a timestamp column based on the transaction data
balances_df['timestamp'] = pd.to_datetime(transactions_df['timestamp'].max())




```

# 1.Calculate net balance changes over time

```python
# Calculate net balance changes over time
transactions_df['net_input'] = transactions_df['input_amounts_x'].apply(lambda x: sum(ast.literal_eval(x)))
transactions_df['net_output'] = transactions_df['output_amounts_y'].apply(lambda x: sum(ast.literal_eval(x)))
transactions_df['net_balance'] = transactions_df['net_output'] - transactions_df['net_input']

# Group by timestamp and aggregate the net balances
transactions_df['timestamp'] = pd.to_datetime(transactions_df['timestamp'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
time_series = transactions_df.groupby('timestamp')['net_balance'].sum().reset_index()

# Filter positive net balances
time_series = time_series[time_series['net_balance'] > 0]

# Plotting the time series data
plt.figure(figsize=(15, 7))
plt.plot(time_series['timestamp'], time_series['net_balance'], label='Net Balance')
plt.xlabel('Timestamp')
plt.ylabel('Net Balance')
plt.title('Net Balance Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

# 2. Wealth Concentration Analysis

```python
# Calculate the total wealth
total_wealth = balances_df['balance'].sum()
# Wealth Concentration Analysis
top_addresses = balances_df.sort_values(by='balance', ascending=False).head(10)
top_addresses_share = top_addresses['balance'].sum() / total_wealth

print(f"Top 10 addresses hold {top_addresses_share:.2%} of the total wealth.")

# Calculate the percentage of total wealth for the top address
top_address_percentage = top_addresses.iloc[0]['balance'] / total_wealth * 100

# Plotting the top addresses' share
plt.figure(figsize=(15, 10))
bars = plt.barh(top_addresses['address'], top_addresses['balance'], color='skyblue')
plt.xlabel('Balance')
plt.ylabel('Address')
plt.title('Top 10 Addresses by Balance')
plt.gca().invert_yaxis()  # Highest balance at the top
plt.grid(True)

# Adding text labels to the bars and annotating the top address percentage
for i, bar in enumerate(bars):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():,.0f}', 
             va='center', ha='left', color='black', fontweight='bold')
    if i == 0:  # Annotate only the top address
        plt.text(bar.get_width() / 2, bar.get_y() + bar.get_height() / 2, 
                 f'{top_address_percentage:.2f}% of total wealth', 
                 va='center', ha='center', color='red', fontweight='bold')

plt.show()
```

# 3. Transaction Volume Analysis

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
from datetime import datetime

# Connect to the SQLite database
conn = sqlite3.connect('iotaDB.db')
query = "SELECT * FROM Transactions"

# Read the data from the database
df = pd.read_sql_query(query, conn)

# Ensure the timestamp column is in datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Check for any NaT values after conversion
print("Number of NaT values in timestamp column:", df['timestamp'].isna().sum())

# Drop rows with NaT values in timestamp column
df = df.dropna(subset=['timestamp'])

# Group by day and count the number of transactions
df['date'] = df['timestamp'].dt.date
transaction_counts = df.groupby('date').size()

# Plot the number of transactions over time
plt.figure(figsize=(12, 6))
transaction_counts.plot(kind='line')
plt.title('Number of Transactions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.grid(True)
plt.show()

# Identify the dates with the highest number of transactions
top_dates = transaction_counts.nlargest(10)
print("Top spike dates and their transaction counts:")
print(top_dates)

# Extract transaction data for the top dates
top_dates_list = top_dates.index.tolist()
spike_transactions = df[df['date'].isin(top_dates_list)]

# Display the transactions that occurred on the top spike dates
print("\nTransactions on top spike dates:")
print(spike_transactions)

# Analyze the transaction data for specific patterns
spike_summary = spike_transactions.groupby(['date', 'transaction_id']).size().reset_index(name='count')

# Display a summary of transactions on spike dates
print("\nSummary of transactions on top spike dates:")
print(spike_summary)
```

# 4. Gini Coefficient Analysis


```python

# Add a timestamp column based on the transaction data
balances_df['timestamp'] = pd.to_datetime(transactions_df['timestamp'].max())

# Function to calculate the Gini coefficient
def gini_coefficient(x):
    x = np.sort(x)
    n = len(x)
    index = np.arange(1, n + 1)
    gini = (np.sum((2 * index - n - 1) * x)) / (n * np.sum(x))
    return gini
```

```python
# Wealth Distribution (by Address)
wealth_gini = gini_coefficient(balances_df['balance'].values)
print(f"Gini Coefficient for Wealth Distribution (by Address): {wealth_gini}")

# Lorenz curve for Wealth Distribution (by Address)
def lorenz_curve(x):
    x = np.sort(x)
    lorenz_curve = np.cumsum(x) / np.sum(x)
    lorenz_curve = np.insert(lorenz_curve, 0, 0)
    lorenz_x = np.linspace(0, 1, len(lorenz_curve))
    return lorenz_x, lorenz_curve

lorenz_x, lorenz_y = lorenz_curve(balances_df['balance'].values)

plt.figure(figsize=(10, 6))
plt.plot(lorenz_x, lorenz_y, label='Lorenz Curve')
plt.plot([0, 1], [0, 1], linestyle='--', color='yellow', label='Line of Ideal Equality')
plt.title('Lorenz Curve for Wealth Distribution (by Address)')
plt.xlabel('Cumulative Share of Addresses')
plt.ylabel('Cumulative Share of Wealth')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Transaction Volume Distribution (by Address)
transactions_df['net_input'] = transactions_df['input_amounts_x'].apply(lambda x: sum(ast.literal_eval(x)))
transactions_df['net_output'] = transactions_df['output_amounts_y'].apply(lambda x: sum(ast.literal_eval(x)))
transactions_df['net_balance'] = transactions_df['net_output'] - transactions_df['net_input']

address_tx_volume = transactions_df.groupby('input_addresses_x')['net_input'].sum().add(
    transactions_df.groupby('output_addresses_y')['net_output'].sum(), fill_value=0)
address_tx_volume = address_tx_volume[address_tx_volume > 0]

tx_volume_gini = gini_coefficient(address_tx_volume.values)
print(f"Gini Coefficient for Transaction Volume Distribution (by Address): {tx_volume_gini}")

# Lorenz curve for Transaction Volume Distribution (by Address)
lorenz_x, lorenz_y = lorenz_curve(address_tx_volume.values)

plt.figure(figsize=(10, 6))
plt.plot(lorenz_x, lorenz_y, label='Lorenz Curve')
plt.plot([0, 1], [0, 1], linestyle='--', color='yellow', label='Line of Ideal Equality')
plt.title('Lorenz Curve for Transaction Volume Distribution (by Address)')
plt.xlabel('Cumulative Share of Addresses')
plt.ylabel('Cumulative Share of Transaction Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Transaction Volume Over Time
transactions_df['month'] = transactions_df['timestamp'].dt.to_period('M')
monthly_tx_volume = transactions_df.groupby('month')['net_balance'].sum()

plt.figure(figsize=(10, 6))
monthly_tx_volume.plot()
plt.title('Transaction Volume Over Time')
plt.xlabel('Month')
plt.ylabel('Transaction Volume')
plt.grid(True)
plt.show()

# Concentration of Large Transactions
large_transactions = transactions_df[transactions_df['net_balance'] > transactions_df['net_balance'].quantile(0.95)]
large_tx_volume = large_transactions.groupby('month')['net_balance'].sum()

plt.figure(figsize=(10, 6))
large_tx_volume.plot()
plt.title('Concentration of Large Transactions Over Time')
plt.xlabel('Month')
plt.ylabel('Transaction Volume')
plt.grid(True)
plt.show()
```
